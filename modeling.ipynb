```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
combinedchis = pd.read_csv("data/combined_chis.csv")
```

```python
# Drop columns with NaN values in them
combinedchis = combinedchis.dropna(axis=1)
combinedchis
```

## Separate input and output variables

```python
X = combinedchis.iloc[:,:-1]
X = X.drop(["PUF1Y_ID"], axis=1)
X
```

```python
y = combinedchis.iloc[:, -1]
y
```

## Split training and test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

## Feature Selection

```python
chis = pd.concat([X_train, y_train], axis=1)
corrmat = chis.corr()
corrmat
```

```python
corrmat["T2D"].nlargest(11)
```

```python
features = corrmat["T2D"].nlargest(11)[1:].index
features
```

```python
# checks if there are any NaN values in any of the features
X_train[features].isnull().sum()
```

```python
# X_train = X_train[features].dropna(axis=1)
X_train = X_train[features]
X_train
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=0)
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix = confusion_matrix(y_test, y_pred)
print(accuracy_score(y_test, y_pred))
print(confusion_matrix)
```
